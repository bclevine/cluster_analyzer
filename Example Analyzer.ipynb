```python
def load_file(filename, MASKING=True):
    print('Reading', filename, 'into DataFrame...')
    with open(filename, 'rb') as f:
        red_rich = np.load(f, allow_pickle=True)
    red_rich = red_rich[0]
    print('Length:', len(red_rich))
    
    #This is sorta tedious but you have to update this function with all the values you're returning
    #At some point maybe I'll make this more automatic...
    
    richlist = []
    redlist = []
    mean_redlist = []
    ilist = []
    mask_frac = []
    mag_r = []
    mag_z = []
    ralist = []
    declist = []
    initialz = []

    for i in red_rich:
        try:
            if i != None:
                if MASKING:
                    richlist.append(i[0])
                    redlist.append(i[1])
                    mean_redlist.append(i[2])
                    ilist.append(i[3])
                    mask_frac.append(i[4])
                    ralist.append(i[5])
                    declist.append(i[6])
                    mag_r.append(i[7])
                    mag_z.append(i[8])
                    initialz.append(i[9])
                else:
                    richlist.append(i[0])
                    redlist.append(i[1])
                    mean_redlist.append(i[2])
                    ilist.append(i[3])
                    ralist.append(i[4])
                    declist.append(i[5])
                    mag_r.append(i[6])
                    mag_z.append(i[7])
                    initialz.append(i[8])
        except:
            if MASKING:
                richlist.append(i[0])
                redlist.append(i[1])
                mean_redlist.append(i[2])
                ilist.append(i[3])
                mask_frac.append(i[4])
                ralist.append(i[5])
                declist.append(i[6])
                mag_r.append(i[7])
                mag_z.append(i[8])
                initialz.append(i[9])
            else:
                richlist.append(i[0])
                redlist.append(i[1])
                mean_redlist.append(i[2])
                ilist.append(i[3])
                ralist.append(i[4])
                declist.append(i[5])
                mag_r.append(i[6])
                mag_z.append(i[7])
                initialz.append(i[8])
    richlist = np.array(richlist)
    redlist = np.array(redlist)
    mean_redlist = np.array(mean_redlist)
    ilist = np.array(ilist)
    mask_frac = np.array(mask_frac)
    mask_frac = mask_frac/.678
    ralist = np.array(ralist)
    declist = np.array(declist)
    mag_r = np.array(mag_r)
    mag_z = np.array(mag_z)
    initialz = np.array(initialz)
    
    return pd.DataFrame({
        'i':ilist,
        'richness':richlist,
        'redshift':redlist,
        'mean_z':mean_redlist,
        'mask_frac':mask_frac,
        'mag_r':mag_r,
        'mag_z':mag_z,
        'ra':ralist,
        'dec':declist,
        'delta_z':redlist-initialz
    })
```
