```python
def rep_nones(arr, n_tiles):
    try:
        if arr.count(None) == len(arr):
            return np.repeat(None, n_tiles)
        else:
            return np.asarray(arr)
    except:
        return np.repeat(None, n_tiles)
    
def load_file(filename, columns):
    print('Reading', filename, 'into DataFrame...')
    with open(filename, 'rb') as f:
        red_rich = np.load(f, allow_pickle=True)
    red_rich = red_rich[0]
    print('Length:', len(red_rich))
    
    temp_arr = []
    
    for i in red_rich:
        temp_arr.append(rep_nones(i, len(columns)))
            
    output_cat = pd.DataFrame(temp_arr, columns=columns)
    return output_cat
```
